## 1. numpy의 활용 

**numpy**는 *ndarray*라는 형식을 활용해서 다양한 행렬 연산과 처리를 지원


### 1.1. ndarray 생성 

```python
import numpy as np
```

```python
# 리스트 형식을 ndarray 형식으로 변환하기
x = [1, 2, 4, 8]
y = np.array(x)
y
```

```python
# 형식 확인
type(y)
```

```python
# 수열 생성
    ## 0부터 10까지 101개 숫자
np.linspace(0, 10, 101)
```

```python
# 랜덤 숫자 뽑기
np.random.randint(1, 46, 6)
```

```python
# 중첩(nested) 리스트를 활용한 다차원 행렬 만들기
y2 = np.array([ [1,2,3,4], [9,8,7,6] ])
y3 = np.array([ [0, 1], [1, 0] ])

y2, y3              
```

```python
# 행렬 크기 확인
y2.shape
    ## 2x4, 2행 4열, 2개 관측치 4개 변수
```

<br>

### 1.2. ndarray와 연산 

numpy에는 ndarray에 적용할 수 있는 다양한 연산자와 함수가 준비되어 있음

```python
# 산술 연산
y2 * 2 + 1
```

```python
# 행렬 곱
y3@y2
```

```python
# 전체 합계
np.sum(y2)
```

```python
# 열 합계
np.sum(y2, axis=0)
```

```python
# 행 평균
np.mean(y2, axis=1)
```

<br>

### 1.3. 부분 선택
대괄호 인덱스를 활용하여 ndarray에서 일부를 선택할 수 있습니다. 

```python
y
```

```python
# 특정 인덱스 선택
y[1]
```

```python
# 슬라이스 활용
y[1:]
```

```python
y[:2]
```

```python
y[:-1]
```

```python
# 다차원 행렬의 부분 선택
    # 콤마 앞은 행, 뒤는 열
y2[:, :]
```

```python
y2[1, 1:3]
```

### 1.4. 길이와 거리 계산

ndarray를 활용해서 길이나 관측치 간 거리를 계산 가능

```python
a = np.array([1,3])
b = np.array([4,3])
```

```python
# 길이(norm) 계산
# linalg : linear algebra
np.linalg.norm(a), np.linalg.norm(b)
```

```python
# 거리 계산
print(np.linalg.norm(a-b))
```

### 2.2. 변수 선택 

대괄호 인덱스에 변수 이름을 입력하고 변수를 선택 가능  
리스트로 변수 이름을 묶고 여러 변수를 한번에 선택 가능

```python
# 한 변수 선택 
    # Series 형식으로 출력됨
df_ins['age']
```

```python
# 한 변수 선택 
    # DataFrame 형식으로 출력됨
df_ins[['age']]
```

### 2.3. 관측치 부분 선택

특정 변수를 선택한 다음 조건을 설정하고 실행하면 True/False로 구성된 bool Series가 계산됨
bool Series를 데이터 대괄호 인덱스에 넣어 조건과 일치하는 관측치를 선택 가능

```python
# 변수 선택
df_ins['age']
```

```python
# 조건 설정
df_ins['age'] < 30
```

```python
# 조건과 일치하는 관측치 선택
df_ins[df_ins['age'] < 30]
```

```python
# 일치 여부를 활용한 특정 그룹 선택
df_ins_male = df_ins[ df_ins['sex'] == 'male' ] 
df_ins_male
```
